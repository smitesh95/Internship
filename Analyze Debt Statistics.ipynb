# 4. Totaling the amount of debt owed by the countries

```python
import numpy
df=pd.read_excel('CountriesDebtdata.xlsx',sheet_name='Data')
df.head(10)
```

```python
#Debt owned by all countries
data=df.iloc[:,6:].fillna(0).sum()
grandtotal=data.astype(numpy.int64).sum()
print(grandtotal)
```

```python
#Debt owned by specific country
def totaldebt(data):
    dist_contry=data['Country Name'].unique()
    for i in dist_contry:
        datanew=data[data['Country Name']==i]
        total=datanew.iloc[:,6:].fillna(0).sum() #fill null alues with 0
        grandtotal=total.astype(numpy.int64).sum() #converted float values to int and sum them
        print( 'Total debt of {} is {}'.format(i,grandtotal))
totaldebt(df)
```

# 5. Country with the highest debt

```python
dict={}
def totaldebt(data):
    dist_contry=data['Country Name'].unique()
    for i in dist_contry:
        datanew=data[data['Country Name']==i]
        total=datanew.iloc[:,6:].fillna(0).sum() #fill null alues with 0
        grandtotal=total.astype(numpy.int64).sum() #converted float values to int and sum all value 
        dict[str(i)]=grandtotal #added data in key and value form in dictionary
    countrymaxdebt=[country for country ,debtvalue in dict.items() if debtvalue==max(dict.values())] #we get the key i.e country name as we set the condition in value part i.e max amount in debtvalue variable
    print('Country with maximun debt is {} with amount {} '.format(countrymaxdebt[0],max(dict.values())))
    
totaldebt(df)
```

# 6. Average amount of debt across indicators

```python
def AverageDebt(data):
    for i in data.iloc[:,6:].columns:#Took the columns range
        data[i]=data[i].fillna(0) #fill null values with zero
    data1=data.groupby('Series Code').sum()#group by and sum the values
    for i in data1.index: #took the index 
        print('Average amount of debt indicator for {} is {} '.format(i,data1.loc[i].mean()))#Findout the mean values bases on index i.e indicator code
AverageDebt(df)
df=pd.read_excel('CountriesDebtdata.xlsx',sheet_name='Data')
```

# 7. The highest amount of principal repayments

```python
def AverageDebt(data):
    amt=[]
    code=[]
    for i in data.iloc[:,6:].columns:#Took the columns range
        data[i]=data[i].fillna(0)#Fill null values with zero
    data1=data.groupby('Series Code').sum()#groupby and sum the values on bases of code
    for i in data1.index:#Took the index 
        data2=data1.loc[i].mean()#Loc work on index so passed the index value and calculated mean
        amt.append(data2)#appended the mean amount in empty list
        code.append(i)#appended the index value in empty list
    count=max(amt)#took max amount from the list
    print(count)#Printed max value
    index=amt.index(count)#Foundout the index value of max amount
    print("Higesht amount of principle reatyment will be done by the max debt indicator code : ",code[index])#on bases of index found the code value from code list
    #print(amt)
    #print(code)
AverageDebt(df)#pass the dataframe to function
df=pd.read_excel('CountriesDebtdata.xlsx',sheet_name='Data')
```
